```python
people = ['leena','lekya','niharika','nithya','thrishala','vicky']
age = [21,18,19,17,20,22]
weight = [65,56,70,50,52,63]
height = [170,155,175,160,150,149]
```

```python
import matplotlib.pyplot as plt #creating static,animated and interactive visualizations
```

```python
plt.scatter(weight,height)
plt.title("Relationship between weight and height of patients")
plt.ylabel("height in meters")
plt.xlabel("weight of patients in kgs")
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.title("People's weight in kgs",fontsize=15,
          fontstyle='italic')
#main plot function 'bar'
plt.bar(x=people,height=weight,width=0.6,
        color='black',edgecolor='k',alpha=0.6)
#edgecolor='k': sets the edge color of the bars to black
#alpha=0.6: sets the transparency of the bars, with 1 being fully opaque and 0 fully transparent
plt.xlabel("People",fontsize=15)
plt.ylabel("Weight (in kgs)",fontsize=15)
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(5,5))
plt.hist(weight,color='blue',edgecolor='k',alpha=0.75,bins=5)
#bins=5: the no.of bins(intervals) to divide the data range into.
plt.title("Histogram of patient weight",fontsize=18)
plt.xlabel("weight in kgs",fontsize=15)
plt.show()
```
